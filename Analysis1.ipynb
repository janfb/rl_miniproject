```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import Maze
import analysisTools as tools
```

Figure for comparison of learning curves for different lambdas

```python
Ntr = 200
l = np.arange(0, 1, .1)
lats = np.zeros((len(l), Ntr))
for i,lam in enumerate(l): 
    m = Maze.Maze(lambda_=lam)
    m.run(Ntr, 5, False)
    lats[i] = np.array(m.get_learning_curve())
```

```python
ymax = lats.max()
plt.figure(figsize=(20,8))
for i in range(len(l)): 
    plt.subplot(2,5,i+1)
    plt.plot(lats[i])
    plt.title("lambda=%.2f"%(l[i]))
    if i>4: 
        plt.xlabel("trials")
plt.subplot(251)
plt.ylabel("Escape latency")
plt.subplot(256)
plt.ylabel("Escape latency")
plt.savefig("figures/learning_curve_lambda.png")
```

```python
Ntr = 200
actions = [4, 8, 16]
lats = np.zeros((len(actions), Ntr))
for i,a in enumerate(actions): 
    m = Maze.Maze(lambda_=0.5, Nactions=a)
    m.run(Ntr, 5)
    lats[i] = m.get_learning_curve()
```

```python
plt.figure(figsize=(20, 5))
for i in range(len(actions)): 
    plt.subplot(1,3,i+1)
    plt.plot(lats[i])
    plt.title("#actions={}".format(actions[i]))
plt.savefig("figures/lc_act.png")
```
